- Get outliers data and without outliers data 


```python
q1=np.percentile(wage_data,25)
q2=np.percentile(wage_data,50)
q3=np.percentile(wage_data,75)
iqr=q3-q1
lb=q1-1.5*iqr
up=q3+1.5*iqr
cond1=wage_data<lb
cond2=wage_data>up
main_con=cond1|cond2
outlier_data=visa_df[main_con]
outlier_df=pd.DataFrame(outlier_data)
outlier_df
```

```python
q1=np.percentile(wage_data,25)
q2=np.percentile(wage_data,50)
q3=np.percentile(wage_data,75)
iqr=q3-q1
lb=q1-1.5*iqr
up=q3+1.5*iqr
cond1=wage_data>lb
cond2=wage_data<up
main_con=cond1 & cond2
non_outlier_data=visa_df[main_con]
non_outlier_data
```
