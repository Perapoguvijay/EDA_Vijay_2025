**Log Transformations**

- log transformation means perform the logrithms operation on original data

- It is one apporach to convert data to normally

- log means naturaally logarithm base=e


```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data)
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1).hist(exp_data,bins=100,label='Exp_Data')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=100,label='Log_Data')
plt.legend()
plt.show()
```

**Power Transfortion**

- power transformation is used to reduce the skewness,so that distribution become symmertric 

    - under these two types are there 

        - Box-cox transformation 

        - Its apply for only positive data

        - Its has both log transformations and sqrt transformations 

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
exp_data=exp_data.reshape(-1,1)
trans_exp=pt.fit_transform(exp_data)
plt.hist(trans_exp,bins=100,label='TRanspose_exp')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='box-cox')
exp_data=exp_data.reshape(-1,1)
trans_data=pt.fit_transform(exp_data)
plt.hist(trans_data,bins=100,label='Trans_data')
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='box-cox')
exp_data=exp_data.reshape(-1,1)
trans_exp=pt.fit_transform(exp_data)
trans_sqrt=np.sqrt(trans_exp)
plt.hist(trans_sqrt,bins=100,label='Sqrt_trans')
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
exp_data=exp_data.reshape(-1,1)
trans_exp=pt.fit_transform(exp_data)
trans_log=np.log(trans_exp)
plt.hist(trans_log,bins=100,label='Log_trans')
plt.show()
```
