**step-4:Modify the Index**

```python
names=['Ramesh','Suresh','Mahesh']
age=[20,21,22]
city=['bglr','chennai','hyd']
cols=['Names','Age','city']
idx=['A','B','C']
pd.DataFrame(zip(names,age,city),index=idx,columns=cols)
```
