```python
continent=visa_df['continent'].value_counts().keys()
continent
```

```python
count=visa_df['continent'].value_counts().values
count
```

```python
plt.bar(continent,count)
```

```python
class_names=visa_df['continent'].value_counts().keys()
class_freq=visa_df['continent'].value_counts().values

colors=['Red','Green','Yellow','Orange','Blue','Cyan']
plt.figure(figsize=(10,5))
plt.bar(class_names,class_freq,color=colors)
plt.xlabel('Class Names')
plt.ylabel('Class Frequency')
plt.title('Continent_Barchart')
plt.savefig('continent_Barchart.jpg')
plt.show()
```
