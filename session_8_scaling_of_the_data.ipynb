- z = (x-μ)/σ

- The Z-score formula is: Z = (x - μ) / σ, where: 
- Z = The Z-score
- x = The individual data point or value
- μ = The population mean
- σ = The population standard deviation.

```python
#step-1 x=prevailaind wage data
#step-2:calculate :u
#step-3:Calculated std:sigma 
#step-4: Nr=step1-step2
#step-5: z=step4/step3
#step-6:create a newc 
```

```python
wage_data=visa_df['prevailing_wage']
u=np.mean(wage_data)
standard_dev=np.std(wage_data)
net_rate=u-standard_dev
net_rate/standard_dev
visa_df['prevailing_wage_z']=net_rate/standard_dev
visa_df
```

```python
x=visa_df['prevailing_wage']
mean_d=x.mean()
mstd=x.std()
nr=x-mean_d

visa_df['p_wage_2']=nr/mstd
visa_df[['prevailing_wage','p_wage_2']]
```

```python
plt.figure(figsize=(8,3))
plt.subplot(1,2,1).hist(visa_df['prevailing_wage'])
plt.subplot(1,2,2).hist(visa_df['p_wage_2'])
plt.show()
```

**Normalization**

```python
x`=(x - x_min) / (x_max - x_min). 
```

```python
wage_data_x=visa_df['prevailing_wage']
x=wage_data_x.mean()
x_min=wage_data_x.min()
x_max=wage_data_x.max()
Nr=wage_data_x-x_min
Dr=x_max-x_min
wage_norm=Nr/Dr
visa_df['p_wage_norm']=Nr/Dr
```
